# Problem 1-a: K-means

```python
def kmeans(data, k=2, iteration=10):
    """
    data has shape (D, N)
    D is the dimension of the data
    N is the number of data
    """
    D, N = data.shape
    # Pick k random data as initial center
    data_idx = np.arange(N)
    centers_idx = np.random.choice(data_idx, k, replace=False)
    
    # Centers has shape (k, D)
    # k is the number of centers
    # D is the dimension
    centers = data[:, centers_idx].T
    
    groups = []
    for i in range(iteration):
        # Row i in dist is the distance between data and center i
        dist = np.sum((data - centers.reshape(k, 2, 1))**2, axis=1)
        # Find closest center for data
        label = np.argmin(dist, axis=0)
        groups = []
        centers = []
        for group_label in range(k):
            # Group data with same closest center
            group = data[:, np.where(label==group_label)[0]]
            groups.append(group)
            # Calculate new mean
            group_center = np.mean(group, axis=1)
            centers.append(group_center)
        centers = np.array(centers)
    return groups
```

```python
def draw_clusters(title, clusters, ax, color="tab20"):
    ax.set_title(title)
    ax.set_aspect('equal')
    # Map distinct color to index
    cmap = plt.cm.get_cmap(color, len(clusters))
    for i, cluster in enumerate(clusters):
        ax.scatter(cluster[0, :], cluster[1, :], c=np.array([cmap(i)]))             
```

```python
np.random.seed(45)
data1_clusters_2_kmeans = kmeans(data1_raw, k=2, iteration=10)
data1_clusters_4_kmeans = kmeans(data1_raw, k=4, iteration=3)
data1_clusters_8_kmeans = kmeans(data1_raw, k=8, iteration=3)
```

# Problem 1-b: Spectral Clustering

```python
def spectral_clustering(data, sigma=10):
    """
    data has shape (D, N)
    D is the dimension of the data
    N is the number of data
    """
    _, N = data.shape
    # Similarity matrix W
    W = np.zeros((N, N))
    D_diag = np.zeros((N,))
    for i in range(N):
        x_i = data[:, i]
        for j in range(N):
            # Skip the diagonal element
            if i==j:
                continue
            # Use the symmetry property 
            if i>j:
                W[i, j] = W[j, i]
                continue
            x_j = data[:, j]
            # Calculate similarity (Gaussian kernel)
            norm = np.sqrt(np.sum((x_i - x_j)**2))
            W[i, j] = np.exp(- norm / sigma)
        D_diag[i] = np.sum(W[i, :])
        
    # Diagonal matrix
    D = np.diag(D_diag)
    # Graph laplacian
    L = D-W
    D_root = np.diag(D_diag**(-0.5))
    normalized_L = D_root.dot(L).dot(D_root)
    eig, vec = np.linalg.eig(normalized_L)
    # Find second smallest eigenvalue
    # The smallest will be zero
    second_small_idx = np.argsort(eig)[1]
    class1 = data[:, np.where((vec[:, second_small_idx] > 0))[0]]
    class2 = data[:, np.where((vec[:, second_small_idx] <= 0))[0]]
    
    return class1, class2 
```

```python
def spectral_clustering_iter(data, iteration=3, sigma=10):
    clusters = [data]
    for i in range(iteration):
        child_clusters = []
        # Cluster every cluster in previous iteration
        for cluster in clusters:
            child_clusters.extend(spectral_clustering(cluster, sigma))
        clusters = child_clusters
    return clusters
```

```python
data1_clusters_2_spectral = spectral_clustering_iter(data1_raw, iteration=1)
data1_clusters_4_spectral = spectral_clustering_iter(data1_raw, iteration=2)
data1_clusters_8_spectral = spectral_clustering_iter(data1_raw, iteration=3)
```

```python
def cluster_qulity_criterion(clusters):
    J = 0
    for cluster in clusters:
        cluster_cost = 0
        for i in range(len(cluster)):
            dist_vec = (cluster - cluster[:, [i]])**2
            dist = np.sqrt(np.sum(dist_vec, axis=0))
            cluster_cost += np.sum(dist)
        J += cluster_cost / len(cluster)
    J /= 2
    return J
```

```python
import pandas as pd
# data1 cluster quaility compare
data1_clusters_2_kmeans_J = cluster_qulity_criterion(data1_clusters_2_kmeans)
data1_clusters_2_spectral_J = cluster_qulity_criterion(data1_clusters_2_spectral)

data1_clusters_4_kmeans_J = cluster_qulity_criterion(data1_clusters_4_kmeans)
data1_clusters_4_spectral_J = cluster_qulity_criterion(data1_clusters_4_spectral)

data1_clusters_8_kmeans_J = cluster_qulity_criterion(data1_clusters_8_kmeans)
data1_clusters_8_spectral_J = cluster_qulity_criterion(data1_clusters_8_spectral)

cost_table = np.array([[data1_clusters_2_kmeans_J, data1_clusters_4_kmeans_J, data1_clusters_8_kmeans_J], 
                       [data1_clusters_2_spectral_J, data1_clusters_4_spectral_J, data1_clusters_8_spectral_J]])
cost_table_df = pd.DataFrame(cost_table, columns = ['2 clusters', '4 clusters', '8 clusters'])
cost_table_df = cost_table_df.rename(index={0:'K-means', 1:'Spectral Clustering'})

cost_table_df = cost_table_df.style.set_caption('Cost comparison on data1')
display(cost_table_df)
```

# Problem 2-g: PCA vs. Kernel PCA

## PCA

```python
# Find eigenvalue and eigenvector of covariance matrix
data1_cov = np.cov(data1_raw)
eig, vec = np.linalg.eig(data1_cov)
eig_ordered_idx = np.argsort(eig)
print("Maximum eigenvalue: ", eig[eig_ordered_idx[-1]])
max_eig_vec = vec[:, eig_ordered_idx[-1]]
print("Eigenvector for maximum eigenvalue: ", max_eig_vec)
```

```python
data1_reduced = max_eig_vec.dot(data1_raw)
print("data1 shape after PCA: ", data1_reduced.shape)
```

```python
fig, ax = plt.subplots()
ax.set_title("Histogram for data1 after PCA")
ax.hist(data1_reduced, 20)
```

## Kernel PCA

```python
def kernel(x, y, sigma=50):
    """
    x and y should be row vectors
    """
    if len(x.shape)==1:
        x = x.reshape((1, -1))
    # Gaussian kernel
    dist_sq = np.sum((x - y)**2, axis=1)
    return np.exp(-dist_sq / sigma)

def kernel_matrix(data, sigma=50):
    """
    data has shape (D, N)
    D is the feature dimension
    N is the number of data
    """
    D, N = data.shape
    data_T = data.T
    K = np.zeros((N, N))
    for i in range(N):
        K[i, :] = kernel(data_T, data_T[i], sigma)
        
    return K

def kernel_PCA(data, n_component=1, sigma=50):
    """
    data should be centered
    data has shape (D, N)
    D is the feature dimension
    N is the number of data
    """
    D, N = data.shape
    # Create kernel matrix
    K = kernel_matrix(data, sigma)
    
    # Modify K for non-centered data
    one_n = np.ones((N,N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n) 
    
    # Kernel matrix is symmetry use eigh instead of eig
    eig, vec = np.linalg.eigh(K)
    
    # Find eigenvectors of max n eigenvalues
    eig_ordered_idx = np.argsort(eig)
    principle_components = [vec[:, eig_ordered_idx[-1]] for i in range(n_component)]
    
    prnciple_components = np.stack(principle_components)
    return prnciple_components
```

```python
data1_reduced_kernel = kernel_PCA(data1_raw, n_component=1, sigma=50)[0]
```

```python
fig, ax = plt.subplots()
ax.set_title("Histogram for data1 after kernel PCA")
ax.hist(data1_reduced_kernel, 20)
```
